### Vectorization

```python
arr =  np.array([-1,-3,4,5,0, -2])
```

```python
arr
```

```python
def classify(a):
    if a<0:
        return 'neg'
    elif a==0:
        return 'zero'
    else:
        return 'pos'
```

```python
new_l = []
for i in arr:
    new_l.append(classify(i))
print(new_l)
```

```python
classifier = np.vectorize(classify)
```

```python
print(classifier(arr))
```
